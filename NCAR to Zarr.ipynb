```python
import xarray as xr
import os
from dask.diagnostics import ProgressBar
# import zarr # No compute image with both xarray and zarr
```

```python
folder_path = '/home/idies/workspace/turb/data02_02/ariel-6-timestep-ncar'

# Loop over files in the folder
for file_name in os.listdir(folder_path):
    if os.path.isfile(os.path.join(folder_path, file_name)):
        # Perform operations on each file

        # https://github.com/pangeo-data/pangeo/issues/150
        # with dask.config.set(get=dask.threaded.get):
        with ProgressBar():
            data_xr = xr.open_dataset(file_name)
            # Set chunk size
            chunk_size = {"nnx": 64, "nny": 64, "nnz": 64}

            data_xr.chunk(chunk_size)
            
            # Disable compression
            encoding = {variable_name: {'compressor': None} for variable_name in data_xr.variables}

            # overwrite if exists
            data_xr.to_zarr(store="/home/idies/workspace/turb/data02_02/" + file_name.replace(".", "-") + "-zarr",
                            mode="w",
                           encoding = encoding)
        
        print(file_name)
```
